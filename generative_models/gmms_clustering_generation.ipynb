```python
from utils import load_iris_dataset, display_first_n_rows, plot_scatter
from utils import setup_plotting

# Load the dataset
iris_df = load_iris_dataset()

setup_plotting()
display_first_n_rows(iris_df)
plot_raw = plot_scatter(iris_df, 'sepal_length', 'sepal_width')
```

```python
p_number_clusters = 3
p_covariance_type = 'full'  # {full (default), tied, diag, spherical}
p_max_iter = 10

from utils import GaussianMixture

gmm = GaussianMixture(n_components=p_number_clusters,
                      covariance_type=p_covariance_type,
                      max_iter=p_max_iter)

gmm.fit(iris_df[['petal_length', 'petal_width']]) # estimating (fitting) model parameters using the EM algorithm
gmm_labels = gmm.predict(iris_df[['petal_length', 'petal_width']]) # predicting the labels (clusters) for the samples using the fitted GMM

from utils import plot_gmm_clusters
plot_gmm = plot_gmm_clusters(iris_df, 'petal_length', 'petal_width', gmm_labels, gmm.means_)
```
